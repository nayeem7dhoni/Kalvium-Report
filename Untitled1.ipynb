```python
import pandas as pd
```

```python
lok_sabha_data = {
    'Party': ['BJP', 'INC', 'SP', 'AITC', 'DMK', 'TDP', 'JD(U)', 'SHS', 'AAP'],
    'Seats': [240, 99, 37, 29, 22, 16, 12, 7, 3]
}

df_lok_sabha = pd.DataFrame(lok_sabha_data)

assembly_data = {
    'Party': ['BJP', 'INC', 'SP', 'BSP', 'TMC', 'DMK', 'AIADMK', 'NCP', 'AAP'],
    'Seats': [150, 80, 40, 30, 25, 20, 18, 15, 12]
}

df_assembly = pd.DataFrame(assembly_data)
```

```python
bye_election_data = {
    'Party': ['BJP', 'INC', 'SP', 'BSP', 'TMC', 'AAP'],
    'Seats': [10, 8, 6, 5, 4, 3]
}

df_bye_election = pd.DataFrame(bye_election_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Lok Sabha visualization
plt.figure(figsize=(14, 8))
top_lok_sabha_parties = df_lok_sabha.nlargest(10, 'Seats')
sns.barplot(x='Seats', y='Party', data=top_lok_sabha_parties, palette='viridis')
plt.title('Top Parties by Number of Seats in Lok Sabha Elections')
plt.xlabel('Number of Seats')
plt.ylabel('Party')
plt.show()
```

```python
# Assembly visualization
plt.figure(figsize=(14, 8))
top_assembly_parties = df_assembly.nlargest(10, 'Seats')
sns.barplot(x='Seats', y='Party', data=top_assembly_parties, palette='plasma')
plt.title('Top Parties by Number of Seats in Assembly Elections')
plt.xlabel('Number of Seats')
plt.ylabel('Party')
plt.show()
```

```python
# Bye-elections visualization
plt.figure(figsize=(14, 8))
top_bye_election_parties = df_bye_election.nlargest(10, 'Seats')
sns.barplot(x='Seats', y='Party', data=top_bye_election_parties, palette='inferno')
plt.title('Top Parties by Number of Seats in Assembly Bye-elections')
plt.xlabel('Number of Seats')
plt.ylabel('Party')
plt.show()
```

```python
# Example DataFrame structure for state-wise distribution
statewise_data = {
    'State': ['Uttar Pradesh', 'Maharashtra', 'West Bengal', 'Tamil Nadu', 'Karnataka', 'Gujarat'],
    'BJP': [60, 40, 18, 12, 25, 26],
    'INC': [5, 8, 4, 10, 7, 4],
    'SP': [35, 0, 0, 0, 0, 0],
    'AITC': [0, 0, 22, 0, 0, 0],
    'DMK': [0, 0, 0, 22, 0, 0],
    'Others': [10, 2, 10, 12, 5, 4]
}

df_statewise = pd.DataFrame(statewise_data)

# Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_statewise.set_index('State'), annot=True, cmap='coolwarm')
plt.title('State-wise Distribution of Seats Won by Each Party')
plt.show()
```

```python
# Example data for vote share
vote_share_data = {
    'Party': ['BJP', 'INC', 'SP', 'AITC', 'DMK', 'Others'],
    'Votes': [3000000, 1500000, 1000000, 1200000, 800000, 2000000]
}

df_vote_share = pd.DataFrame(vote_share_data)

# Pie chart
plt.figure(figsize=(10, 8))
plt.pie(df_vote_share['Votes'], labels=df_vote_share['Party'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set3'))
plt.title('Overall Vote Share by Party')
plt.show()
```

```python
# Example trend data over years
trend_data = {
    'Year': [2014, 2019, 2024],
    'BJP': [282, 303, 240],
    'INC': [44, 52, 99],
    'SP': [5, 5, 37],
    'AITC': [34, 22, 29],
    'DMK': [0, 23, 22]
}

df_trend = pd.DataFrame(trend_data)

# Line chart
plt.figure(figsize=(14, 8))
for party in df_trend.columns[1:]:
    plt.plot(df_trend['Year'], df_trend[party], marker='o', label=party)

plt.title('Trends in Seats Won by Parties Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Seats')
plt.legend()
plt.show()
```

```python
# Example box plot data
box_plot_data = {
    'Party': ['BJP', 'BJP', 'BJP', 'INC', 'INC', 'INC', 'SP', 'SP', 'AITC', 'AITC', 'DMK', 'DMK'],
    'State': ['UP', 'MH', 'KA', 'UP', 'MH', 'KA', 'UP', 'MH', 'WB', 'WB', 'TN', 'TN'],
    'Seats': [60, 40, 25, 5, 8, 7, 35, 0, 22, 22, 22, 22]
}

df_box_plot = pd.DataFrame(box_plot_data)

# Box plot
plt.figure(figsize=(14, 8))
sns.boxplot(x='Party', y='Seats', data=df_box_plot)
plt.title('Distribution of Seats Won by Each Party Across States')
plt.show()
```

```python
# Example scatter plot data
scatter_data = {
    'Party': ['BJP', 'INC', 'SP', 'AITC', 'DMK'],
    'Seats': [240, 99, 37, 29, 22],
    'Voter Turnout (%)': [67.1, 64.5, 62.3, 69.2, 68.0]
}

df_scatter = pd.DataFrame(scatter_data)

# Scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Voter Turnout (%)', y='Seats', hue='Party', size='Seats', data=df_scatter, palette='muted', sizes=(100, 200))
plt.title('Correlation Between Seats Won and Voter Turnout')
plt.xlabel('Voter Turnout (%)')
plt.ylabel('Number of Seats')
plt.show()
```
